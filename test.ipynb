```python
import rasterio as rio
import numpy as np

with rio.open(r'clipped_raster\2004_2009_clipped.tif') as src:
    r1 = src.read(1)
    print(np.unique(r1))

with rio.open(r'clipped_raster\2020_2045_RCP45_clipped.tif') as src:
    r2 = src.read(1)
    print(np.unique(r2))
```

```python
import matplotlib.pyplot as plt
value1 = 0
value2 = 0

# filter the raster to show only the value of interest
r1_f = np.where(r1 == value1, 1, np.nan)
r2_f = np.where(r2 == value2, 1, np.nan)

plt.imshow(r1_f)
plt.show()

plt.imshow(r2_f)
plt.show()
```
